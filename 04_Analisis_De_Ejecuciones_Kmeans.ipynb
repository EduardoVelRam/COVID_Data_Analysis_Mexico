```python
# Definición de Funciones 

def vert_filt(dframe, valores):
    nombre = dframe[valores]
    print('Número de atributos del dataframe: ' + str(len(valores)) + '; cada uno con ' + str(len(nombre)) + ' datos.')
    return nombre

# Fitro horizontal
def hor_filt(dframe, var, restriccion, valor):
    if restriccion == '=':
        indices = dframe[dframe[var] == valor].index
        nuevo = dframe.drop(indices)
        print('Longitud nueva del dataframe: '+str(len(nuevo))+'. Filtro de '+str(var)+str(restriccion)+str(valor)+' aplicado.')
    if restriccion == '>':
        indices = dframe[dframe[var] > valor].index
        nuevo = dframe.drop(indices)
        print('Longitud nueva del dataframe: '+str(len(nuevo))+'. Filtro de '+str(var)+str(restriccion)+str(valor)+' aplicado.')
    if restriccion == '<':
        indices = dframe[dframe[var] < valor].index
        nuevo = dframe.drop(indices)
        print('Longitud nueva del dataframe: '+str(len(nuevo))+'. Filtro de '+str(var)+str(restriccion)+str(valor)+' aplicado.')
    return nuevo

def vert_filt2(dframe,atributo, claves):
    nuevo = dframe[dframe[atributo].isin(claves)]
    return nuevo

def acople_col(dframe, lista, nombre): 
    loc_n = dframe.shape[1]
    dframe.insert(
        loc = loc_n,        
        column = nombre,
        value = lista,
        allow_duplicates = False, 
    )
    
def str_conv(dframe, atributo):  # PARA UN DATAFRAME
    lista = list(dframe[atributo].values)
    col_nueva = []
    for i in range(len(dframe)):
        col_nueva.append(str(lista[i]))
    return col_nueva
        
def str_conv2(lst):   # PARA UNA LISTA
    col_nueva = []
    for i in range(len(lst)):
        col_nueva.append(str(lst[i]))
    return col_nueva

# Función que promedie la mortalidad por membresía
def prom_mort_mem(dframe, atributo,memb, grupo):  # atributo debe ser mortalidad_norm o mortalidad_norm_21 y grupo debe ser una lista
    nuv = vert_filt2(dframe, memb, grupo)
    lista_nuv = list(nuv[atributo])
    prom = sum(lista_nuv)/len(nuv)
    return prom
    
def lista_morts(dframe, atributo,memb, grupos):  
    listam = []
    for i in range(0,grupos):
        mort = prom_mort_mem(dframe, atributo,memb, [i])
        listam.append(round(mort,4)) # para cuando quiera los máximos y mínimos
    listam.sort()
    return listam
    
def lista_morts_label(dframe, atributo,memb, grupos):  
    listam = []
    for i in range(0,grupos):
        mort = prom_mort_mem(dframe, atributo,memb, [i])
        #listam.append(mort) # para cuando quiera los máximos y mínimos
        listam.append('Grupo ' + str(i) + ': ' + str(round(mort,4)))
    return listam

# FUnción que devuelva extremos
def extremos(dframe, atributo,memb, grupos):
    exts = []
    for i in range(0,grupos):
        mort = prom_mort_mem(dframe, atributo,memb, [i])
        exts.append(str(round(mort,4))+'-'+str(i))
    exts.sort()
    exts = exts[0:3]+exts[13:16]
    # Puede incorporarse una función que arroje una lista únicamente de los extremos
    return exts
```
