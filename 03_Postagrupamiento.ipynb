```python
# Definición de Funciones 

def vert_filt(dframe, valores):
    nombre = dframe[valores]
    print('Número de atributos del dataframe: ' + str(len(valores)) + '; cada uno con ' + str(len(nombre)) + ' datos.')
    return nombre

# Fitro horizontal
# dframe: el dataframe de interés, var: atributo, restriccion: tipo de restriccion (=, <, >) y valor: valor de la restricción
def hor_filt(dframe, var, restriccion, valor):
    if restriccion == '=':
        indices = dframe[dframe[var] == valor].index
        nuevo = dframe.drop(indices)
        print('Longitud nueva del dataframe: '+str(len(nuevo))+'. Filtro de '+str(var)+str(restriccion)+str(valor)+' aplicado.')
    if restriccion == '>':
        indices = dframe[dframe[var] > valor].index
        nuevo = dframe.drop(indices)
        print('Longitud nueva del dataframe: '+str(len(nuevo))+'. Filtro de '+str(var)+str(restriccion)+str(valor)+' aplicado.')
    if restriccion == '<':
        indices = dframe[dframe[var] < valor].index
        nuevo = dframe.drop(indices)
        print('Longitud nueva del dataframe: '+str(len(nuevo))+'. Filtro de '+str(var)+str(restriccion)+str(valor)+' aplicado.')
    return nuevo

def vert_filt2(dframe,atributo, claves):
    nuevo = dframe[dframe[atributo].isin(claves)]
    return nuevo

def acople_col(dframe, lista, nombre): # Se solicita el dataframe, la lista de elementos por añadir y el nombre de la columna
    loc_n = dframe.shape[1]
    dframe.insert(
        loc = loc_n,         # Podría buscarse una forma de que lo agregara al final de la lista
        column = nombre,
        value = lista,
        allow_duplicates = False, # False restringe la duplicación de columnas
    )
    
def str_conv(dframe, atributo):  # PARA UN DATAFRAME
    lista = list(dframe[atributo].values)
    col_nueva = []
    for i in range(len(dframe)):
        col_nueva.append(str(lista[i]))
    return col_nueva
        
def str_conv2(lst):   # PARA UNA LISTA
    col_nueva = []
    for i in range(len(lst)):
        col_nueva.append(str(lst[i]))
    return col_nueva

# Función que promedie la mortalidad por membresía
def prom_mort_mem(dframe, atributo, grupo):  # atributo debe ser mortalidad_norm o mortalidad_norm_21 y grupo debe ser una lista
    nuv = vert_filt2(dframe, 'membresias', grupo)
    lista_nuv = list(nuv[atributo])
    prom = sum(lista_nuv)/len(nuv)
    return prom
    
def lista_morts(dframe, atributo, grupos):  
    listam = []
    for i in range(0,grupos):
        mort = prom_mort_mem(dframe, atributo, [i])
        listam.append(round(mort,5)) # para cuando quiera los máximos y mínimos
    listam.sort()
    return listam
    
def lista_morts_label(dframe, atributo, grupos):  
    listam = []
    for i in range(0,grupos):
        mort = prom_mort_mem(dframe, atributo, [i])
        #listam.append(mort) # para cuando quiera los máximos y mínimos
        listam.append('Grupo ' + str(i) + ': ' + str(round(mort,5)))
    return listam
```
