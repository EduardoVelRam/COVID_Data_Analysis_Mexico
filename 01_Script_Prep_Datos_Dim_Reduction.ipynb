```python
import pandas as pd  # sólo necesito una librería de manipulación de datos para esta parte
```

```python
# Subir los datos completos a la carpeta y borrarlos después para no gastar memoria
# Dataset covid
datosCov = pd.read_csv(r'covid_17012023.csv', error_bad_lines=False, engine="python")
#datosCov = pd.read_csv(r'covid_2020.csv', error_bad_lines=False, engine="python")
#datosCov = pd.read_csv(r'covid_2021.csv', error_bad_lines=False, engine="python")
df_Cov = pd.DataFrame(datosCov)
print(df_Cov.head())
# ¿Existe otra forma de hacerlo a menor costo?
```

```python
# Redusco dimensiones del dataset

# Filtor vertical
# Me interesan: 'ENTIDAD_RES', 'MUNICIPIO_RES', 'FECHA_INGRESO', 'FECHA_DEF'*, 'EDAD', 'NACIONALIDAD'*, 
# * Considerar filtrar en este mismo Script


# Aplico filtos verticales

dat_interes_1 = df_Cov[['ENTIDAD_RES', 'MUNICIPIO_RES', 'FECHA_INGRESO', 'FECHA_DEF', 'EDAD', 'NACIONALIDAD']]
#print(dat_interes_1)
#print(type(dat_interes_1['FECHA_DEF']))

# Aplico filtros horizontales

# Ya que analizaré la mortalidad:
# Retiro los que no murieron (Si quisiera analizar casos no la retiro y me quedo con todos)
indices_v = dat_interes_1[dat_interes_1['FECHA_DEF']=='9999-99-99'].index
dat_interes_2 = dat_interes_1.drop(indices_v)
#print(dat_interes_2) # len = 28925

# Retiro los extranjeros
indices_e = dat_interes_2[dat_interes_2['NACIONALIDAD'] != 1].index
dat_interes_3 = dat_interes_2.drop(indices_e)
#print(dat_interes_3) # len = 28851

# Ya que sólo me interesa analizar una entidad en específico la indico:
# Filtro para la CDMX (Si quisiera analizar para todo el país se omite)
indices_mx = dat_interes_3[dat_interes_3['ENTIDAD_RES'] != 9].index # es ENITDAD_UM o ENTIDAD_RES
dat_interes_4 = dat_interes_3.drop(indices_mx)
print(dat_interes_4) # len = 2781
```
